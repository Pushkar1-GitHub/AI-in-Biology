```python
from rdkit import Chem
from rdkit.Chem import rdFingerprintGenerator
```

```python
smiles = 'Cc1onc(-c2ccccc2)c1C(=O)N[C@@H]1C(=O)N2[C@@H](C(=O)O)C(C)(C)S[C@H]12'
mol = Chem.MolFromSmiles(smiles)
```

```python
# Create a Morgan fingerprint generator with a radius of 2 and a size of 2048 bits
morgan_gen = rdFingerprintGenerator.GetMorganGenerator(radius=2, fpSize=2048)

# Generate the fingerprint
morgan_fp = morgan_gen.GetFingerprint(mol)

# Convert to a numpy array if needed for ML
import numpy as np
morgan_fp_array = np.array(morgan_fp)
```

```python
from rdkit.Chem import AllChem
morgan_fp_legacy = AllChem.GetMorganFingerprintAsBitVect(mol, radius=2, nBits=2048)
```

```python
from rdkit.Chem import MACCSkeys

# Generate MACCS keys (no generator needed, uses a direct function)
maccs_fp = MACCSkeys.GenMACCSKeys(mol)

# Convert to a numpy array if needed
maccs_fp_array = np.array(maccs_fp)
```

```python
# Create an RDKit fingerprint generator
rdkit_gen = rdFingerprintGenerator.GetRDKitFPGenerator(fpSize=2048)

# Generate the fingerprint
rdkit_fp = rdkit_gen.GetFingerprint(mol)
```

```python
# Create an Atom Pair generator
ap_gen = rdFingerprintGenerator.GetAtomPairGenerator(fpSize=2048)

# Generate the fingerprint
ap_fp = ap_gen.GetFingerprint(mol)
```
