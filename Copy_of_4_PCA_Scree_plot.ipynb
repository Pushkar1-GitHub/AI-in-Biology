```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# 1. Mock Data (High Dimensional)
# Let's simulate 10 Samples with 20 Genes
# Random data usually has low explained variance, so we add some "structure"
np.random.seed(42)
X = np.dot(np.random.rand(10, 2), np.random.rand(2, 20)) # Structured part
X += 0.1 * np.random.normal(size=(10, 20)) # Add some noise

# 2. Fit PCA
# We generally calculate as many components as min(n_samples, n_features)
pca = PCA().fit(StandardScaler().fit_transform(X))

# 3. Extract Variance
explained_var = pca.explained_variance_ratio_
cumulative_var = np.cumsum(explained_var)

# 4. Plotting
plt.figure(figsize=(10, 6))

# Bar chart for individual components
plt.bar(range(1, len(explained_var)+1), explained_var, alpha=0.5, align='center', label='Individual variance')

# Line chart for cumulative variance
plt.step(range(1, len(cumulative_var)+1), cumulative_var, where='mid', label='Cumulative variance', color='red')

# Formatting
plt.ylabel('Explained Variance Ratio')
plt.xlabel('Principal Component Index')
plt.title('Scree Plot')
plt.legend(loc='best')
plt.grid(True)

# Show the "Elbow" text for teaching purposes
plt.annotate('The Elbow (Cutoff)', xy=(2, explained_var[1]), xytext=(4, 0.4),
             arrowprops=dict(facecolor='black', shrink=0.05))

plt.show()
```
